# Написать функцию, решающую систему линейных уравнений с трёх диагональной матрицей методом Гаусса. Протестировать на случайной матрице размера N = 20

```python
import numpy as np
from random import randint
```

```python
N=int(input())
x = []
b = np.zeros(N)

for i in range(N):
    b[i]=float(randint(1,5))
a=np.zeros((N,N))
for i in range(N):
    a[i,i]=float(randint(1,5))
    if i!=0:
        a[i,i-1]=float(randint(1,5))
    if i!=N-1:
        a[i,i+1]=float(randint(1,5))
print("Исходная матрица")
print(a)
print("Исходные ответы")
print(b)
for i in range(1,N):
    if a[i-1,i-1]!=0:
        k=a[i,i-1]/a[i-1,i-1]
        a[i]-=(a[i-1]*k)
        b[i]=b[i]-b[i-1]*k
    else:
        a[i],a[i-1]=a[i-1],a[i].copy()
        b[i],b[i-1]=b[i-1],b[i].copy()
print("Новая матрица")        
print(a)
print("Cтрочка ответов для проверки")
print(b)
x.append(b[len(a) -  1]/a[len(a)-1][len(a)-1])
for i in range(len(a) - 1, 0, -1):
    x.append((b[i-1] - a[i-1][i]*x[N-i-1])/a[i-1][i-1])
x = x[::-1]

```

```python
print(" Вторая строчка ответов для проверки")
for i in range(N):
    s=0
    for j in range(N):
        s+=(x[j]*a[i,j]) 
    
    print(s, end=' ')
```
