$$x''=-w^2x + kx' - ex^3$$

## Метод Эйлера

```python
import math
import matplotlib.pyplot as plt
from matplotlib import mlab

print( 'w^2=', sep='\n', end='')
w = float(input())
print( 'k=', sep='\n', end='')
k = float(input())
print( 'e=', sep='\n', end='')
e = float(input())
print( 'x_0=', sep='\n', end='')
x_0 = float(input())
print('u_0=', end='')
u_0 = float(input())



# начальные условия
xlist = [x_0]
ulist = [u_0]
 
 
f = lambda x, u: -w*x - k*u + e*x**3
ilist = mlab.frange(0, 100, 0.1)
for i in ilist:
    ulist.append(ulist[-1] + f(xlist[-1], ulist[-1])*(ilist[1] - ilist[0]))
    xlist.append(xlist[-1] + ulist[-2]*(ilist[1] - ilist[0]))

plt.plot( ilist, xlist[1:], '-')


 
    
 
 
```

## Метод Рунге-Кутта


```python
import math
import matplotlib.pyplot as plt
from matplotlib import mlab
 
print( 'w^2=', sep='\n', end='')
w = float(input())
print( 'k=', sep='\n', end='')
k = float(input())
print( 'e=', sep='\n', end='')
e = float(input())
print( 'x_0=', sep='\n', end='')
x_0 = float(input())
print('u_0=', end='')
u_0 = float(input())

# начальные условия
xlist = [x_0]
ulist = [u_0]
 
 
f = lambda x, u: -w*x -k*u + e*x**3
ilist = mlab.frange(0, 100, 0.1)
for i in ilist:
    ulist.append(ulist[-1] + (ilist[1] - ilist[0])*f(xlist[-1] + 0.5*(ilist[1] - ilist[0]), ulist[-1] + 0.5*(ilist[1] - ilist[0])*f(xlist[-1], ulist[-1])))
    xlist.append(xlist[-1] + (ilist[1] - ilist[0])*ulist[-2] )

plt.plot( ilist, xlist[1:], '-')

```
