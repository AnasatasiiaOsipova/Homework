## Метод Эйлера

#### Гармонический осциллятор без затухания

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import mlab

print( 'w^2=', sep='\n', end='')
w = float(input())
print( 'x_0=', sep='\n', end='')
x_0 = float(input())
print('u_0=', end='')
u_0 = float(input())



# начальные условия
xlist = [x_0]
ulist = [u_0]
 
 
f = lambda x, u: -w*x 
ilist = np.linspace(0.0, 10*2*math.pi/(w**0.5), 10**(6))
for i in ilist:
    ulist.append(ulist[-1] + f(xlist[-1], ulist[-1])*(ilist[1] - ilist[0]))
    xlist.append(xlist[-1] + ulist[-2]*(ilist[1] - ilist[0]))
plt.xlabel('t')
plt.ylabel('x')
plt.plot( ilist, xlist[:-1], '-')


```

Проверим закон сохранения энергии
(Так как известна только частота, т.е. отношение некоторых параметров, то получится найти только отнормированную энергию)

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import mlab

print( 'w^2=', sep='\n', end='')
w = float(input())
print( 'x_0=', sep='\n', end='')
x_0 = float(input())
print('u_0=', end='')
u_0 = float(input())



# начальные условия
xlist = [x_0]
ulist = [u_0]
E1 = [w*(x_0)**2]
E2 = [(u_0)**2]
E= [w*(x_0)**2 + (u_0)**2]

f = lambda x, u: -w*x 
ilist = np.linspace(0.0, 10*2*math.pi/(w**0.5), 10**(6))
for i in ilist:
    ulist.append(ulist[-1] + f(xlist[-1], ulist[-1])*(ilist[1] - ilist[0]))
    xlist.append(xlist[-1] + ulist[-2]*(ilist[1] - ilist[0]))
    E1.append(w*(xlist[-1])**2 )
    E2.append((ulist[-1])**2 )
    E.append(E1[-1] + E2[-1])
plt.xlabel('t')
plt.ylabel('E')
plt.plot(ilist, E1[:-1], '-', label='Потенциальная энергия'), plt.plot(ilist, E2[:-1], '-', label='Кинетическая энергия'),  plt.plot(ilist, E[:-1], '-', label='Полная энергия'), plt.legend()
```

Энергия сохраняется
