## numpy exercise 44.

44. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

```python
Z = np.random.random((10,2))
X,Y = Z[:,0], Z[:,1]
R = np.hypot(X, Y)
T = np.arctan2(Y,X)
print(R)
print(T)
```

## A cochleoid curve 

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos
import random

```

```python
%matplotlib notebook
```

```python
a = 1 
b = np.random.random((100))
phi = np.where(b == 0, 1, b)
for p in phi:
    for a in range(1, 10):
        r = a*sin(p)/p 
        plt.plot(r*cos(p), r*sin(p), "o")
```
