### Метод Монте-Карло. Часть 1

```python
import numpy as np
```

```python
import math 
import matplotlib.pyplot as plt
from scipy.stats import norm
import pylab as pl
import numpy.random
```

### Для равномерного распределения

```python
M1 = np.random.uniform(39.95*10**4, 40.05*10**4, 10**6)
M2 = np.random.uniform(29.9*10**4, 30.1*10**4, 10**6)
r = np.random.uniform(3.19, 3.21, 10**6)
F = M1*M2*6.67384*10**(-11)/r/r

plt.hist(F,bins=100, normed = True)


mu, std = norm.fit(F)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)


plt.show()
```

### Для нормального распределения

```python

mu1, sigma1 = 40*10**4, 0.05/3*10**4 
M1 = np.random.normal(mu1, sigma1, 10**6)
mu2, sigma2 = 30*10**4, 0.1/3*10**4 
M2 = np.random.normal(mu2, sigma2, 10**6)
mu3, sigma3 = 3.2, 0.01/3 
r = np.random.normal(mu3, sigma3, 10**6)
F = M1*M2*6.67384*10**(-11)/r/r

plt.hist(F,bins=100, normed = True)

mu, std = norm.fit(F)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)


plt.show()
```

### Метод Монте - Карло.

```python
mu1, sigma1 = 40*10**4, 2/3*10**4 
M1 = np.random.normal(mu1, sigma1, 10**6)
mu2, sigma2 = 30*10**4, 10/3*10**4 
M2 = np.random.normal(mu2, sigma2, 10**6)
mu3, sigma3 = 3.2, 1/3 
r = np.random.normal(mu3, sigma3, 10**6)
F = M1*M2*6.67384*10**(-11)/r/r

plt.hist(F,bins=100, normed = True)

mu, std = norm.fit(F)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)


plt.show()
```
